```python
import numpy as np
from EnsemblePursuit import EnsemblePursuitPyTorch
from scipy.io import loadmat
import numpy as np
from scipy import io
from utils import test_train_split, evaluate_model_torch, subtract_spont
import matplotlib.pyplot as plt
import glob
import os
%matplotlib inline
```

```python
save_path='/home/maria/Documents/EnsemblePursuit/models/'
def fit_n_av_neurons(save_path):
    n_av_neurons=[5,10,15,20,100]
    data = io.loadmat('/home/maria/Documents/EnsemblePursuit/data/natimg2800_M170717_MP034_2017-09-11.mat')
    resp = data['stim'][0]['resp'][0]
    spont = data['stim'][0]['spont'][0]
    istim = (data['stim'][0]['istim'][0]).astype(np.int32)
    X=subtract_spont(spont,resp)
    for n in n_av_neurons:
        neuron_init_dict={'method':'top_k_corr','parameters':{'T':10,'n_of_neurons':1,'min_assembly_size':8,'n_av_neurons':n}}
        ep=EnsemblePursuitPyTorch()
        lambd_=0.03
        U_V,nr_of_neurons,U,V, cost_lst,seed_neurons=ep.fit_transform(X,lambd_,150,neuron_init_dict)
        np.save(save_path+'_'+str(n)+'n_av_neurons.npy',V)
        np.save(save_path+'_'+str(n)+'cost_lst.npy',cost_lst)

fit_n_av_neurons(save_path)
```

```python
model_path='/home/maria/Documents/EnsemblePursuit/models/'
def plot_n_av_cost_lst(model_path):
    fn=[]
    for filename in glob.glob(os.path.join(model_path, '*cost_lst.npy')):
        cost_lst=np.load(filename)
        plt.plot(range(0,150),cost_lst)
        fn.append(filename[46:])
    plt.legend(fn)
    plt.show()

plot_n_av_cost_lst(model_path)
    
```

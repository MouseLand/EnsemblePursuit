```python
from sklearn.metrics import jaccard_similarity_score
import seaborn as sns
import sys
np.set_printoptions(threshold=sys.maxsize)
data_path='/home/maria/Documents/EnsemblePursuit/models/'

def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))

def jaccard_on_ensembles(data_path):
    for filename1 in glob.glob(os.path.join(data_path, '*_neuron_lst.npy')):
        for filename2 in glob.glob(os.path.join(data_path, '*_neuron_lst.npy')):
            if filename1!=filename2:
                print(filename1)
                print(filename2)
                neuron_lst1=np.load(filename1)
                neuron_lst2=np.load(filename2)
                jac_mat=np.zeros((150,150))
                for neurons1 in (0,len(neuron_lst1)-1):
                    for neurons2 in (0,len(neuron_lst2)-1):
                        jac_mat[neurons1,neurons2]=jaccard_similarity(sorted(neuron_lst1[neurons1]),sorted(neuron_lst2[neurons2]))
                print(jac_mat)
                sns.heatmap(jac_mat)
                plt.show()
                
            
                
jaccard_on_ensembles(data_path)
```
